```python
import random
```

```python
class perceptron():
    def __init__(self,noi) -> None:
        self.noi=noi
        self.w=[random.randint(1,3) for i in range(noi+1)]
    
    def get_pred(self,a):
        o=self.w[0]
        for i in range(self.noi):
            o+=a[i]*self.w[i+1]

        return o

    def sgn(self,x):
        if x > 0:
            return 1
        else:
            return -1

    def get_error(self,a,o,t,eta=0.1):
        delta_w=[]
        delta_w.append(eta*(t-o))
        for i in range(1,self.noi+1):
            delta_w.append(eta*(t-o)*a[i-1])

        return delta_w

    def train(self,a,t):
        while True:
            print("Training:\n")
            result=[]
            for i,a1 in enumerate(a): 
                temp=self.get_pred(a1)
                o=self.sgn(temp)
                if(o!=t[i]):
                    delta_w=self.get_error(a1,o,t[i])
                    for i in range(self.noi+1):
                        self.w[i]+=delta_w[i]

                print(a1,o,t[i])
                result.append(o)
            
            print("\n")

            if(result==t):
                break                 
```

```python
a=[[0,0],[0,1],[1,0],[1,1]]
t=[-1,-1,-1,1]
x=perceptron(2)
x.train(a,t)
```
